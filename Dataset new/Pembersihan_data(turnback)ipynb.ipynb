```python
def bersihkan_narasi_berita(teks_berita: str) -> str:
    """
    Membersihkan teks berita untuk mengekstrak bagian 'NARASI:' berdasarkan algoritma yang ditentukan.

    Args:
        teks_berita (str): String teks berita mentah yang mungkin berisi 'NARASI:' dan 'PENJELASAN:'.

    Returns:
        str: Bagian 'NARASI:' yang sudah dibersihkan, atau string kosong jika tidak ditemukan
             atau format tidak sesuai.
    """
    idx_narasi = teks_berita.find("NARASI:")

    # 1. Cek "NARASI:"
    if idx_narasi == -1:  # if false: skip
        return ""

    # Lokasi awal narasi setelah 'NARASI:' dan spasi/newline
    awal_narasi = idx_narasi + len("NARASI:")

    idx_penjelasan = teks_berita.find("PENJELASAN:", awal_narasi)

    # Cek lokasi "PENJELASAN:" terhadap "NARASI:"
    if idx_penjelasan == -1:  # PENJELASAN: tidak ditemukan setelah NARASI:
        # if sebelum: ambil baris setelah "NARASI:" sampai akhir
        # (Logika ini juga mencakup kasus PENJELASAN: tidak ada sama sekali)
        return teks_berita[awal_narasi:].strip()
    else:
        # if sesudah: ambil baris setelah "NARASI:" sampai sebelum "PENJELASAN:"
        return teks_berita[awal_narasi:idx_penjelasan].strip()
```

```python
# prompt: gunakan bersihkan_narasi_berita df

df['Narasi_Bersih'] = df['Isi Artikel'].apply(bersihkan_narasi_berita)
print(df[['Isi Artikel', 'Narasi_Bersih']].head())
```
